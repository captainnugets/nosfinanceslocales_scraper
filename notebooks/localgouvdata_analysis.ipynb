## Analyse des finances des communes sur 2012

 * *source des données financières*: [http://www.collectivites-locales.gouv.fr/](http://www.collectivites-locales.gouv.fr/)
 * *source des codes communes*: [insee](http://www.insee.fr/fr/methodes/nomenclatures/cog/documentation.asp)




### Examinons le nombre de communes crawlées sur le site collectivites-locales.gouv.fr et le nombre de départements concernés.

* On s'attend à avoir récupérer les informations financières de 36 700 communes puisque c'est le nombre de communes déclarées en 2012 ([wikipédia](http://www.insee.fr/fr/methodes/nomenclatures/cog/documentation.asp)). O
* On attend également que ces communes soient réparties dans 101 départements ([insee](http://www.insee.fr/fr/methodes/nomenclatures/cog/documentation.asp))

```python
import json
import os, sys
import pandas as pd
sys.path.append(os.path.abspath('./..'))

from localgouv.spiders.localgouv_spider import uniformize_code, convert_dom_code, convert_city

data_dir = 'scraped_data/'
insee_filepath = 'data/france2013.txt'

quantiles = pd.np.arange(0.01, 1, 0.01)
def plot_quantiles(series, figsize=(12,10)):
    pd.Series(quantiles, index=[series.quantile(q) for q in quantiles]).plot(figsize=figsize)
```

```python
# Load insee code of cities
insee_df = pd.io.parsers.read_csv(insee_filepath, '\t')
```

```python
# Load data on 2012
df = pd.DataFrame(json.load(open(os.path.join(data_dir, 'cities_2012.json'))))
```

```python
print "On récupère les champs suivants pour chacune des communes: \n %s"%df.columns.tolist()
```

```python
# Sanity checks
print u"Nombre de communes dupliquées: %s"%(df['insee_code'].count() - df['insee_code'].unique().size)
print u"Numbre of communes non dupliquées: %s"%df['insee_code'].unique().size
# XXX: error in json data file: insee_code is not defined correctly
df['insee_code_c'] = df['insee_code'].apply(lambda r: r[3:6] + r[0:3])
df['dep'] = df['insee_code_c'].apply(lambda r: r[:3])
gp_by_dep = df.groupby('dep')
print u"Nombre de départements: %s"%gp_by_dep.dep.size().size
```

```python
# Take only current cities (Cf. insee doc http://www.insee.fr/fr/methodes/default.asp?page=nomenclatures/cog/doc_ffrancee.htm)
insee_df['DEP'] = uniformize_code(insee_df, 'DEP')
insee_df['COM'] = uniformize_code(insee_df, 'COM')
insee_df['DEP'] = convert_dom_code(insee_df)
insee_df['COM'] = insee_df.apply(convert_city, axis=1)
insee_df['DEPCOM'] = insee_df['DEP'] + insee_df['COM']
# Remove MAYOTTE department, 
current_insee_df = insee_df[(insee_df['ACTUAL'] == 1) & (insee_df['DEP'] <> '976')] 
print "Nombre de commune sans le département de Mayotte au 2013/01/01 selon l'insee: %s"%current_insee_df['DEPCOM'].unique().size
```

## Observations
 * Un département n'est pas représenté: MAYOTTE (976), ce qui est attendu vu que les données financières des communes de mayotte ne semble pas être présentes sur le site crawlé. Le département de Mayotte compte 17 communes.
 * Le nombre de communes crawlées est de 36663 ce qui est très proche du nombre total de communes en 2012: 36700 - 17 (Mayotte). Il en manque donc 20.
Nous allons donc regarder de plus près si certains budgets de communes seraient absents du site [collectivites-locales.gouv.fr](http://www.collectivites-locales.gouv.fr/)

## Analyse des taxes d'habitation, taxes foncières, charges de personnel et ratio d'endettement

On s'intéresse dans cette partie à la qualité de quelques informations financières: on commence par des informations relativement bien connues du grand public, et donc simple à appréhender.

```python
# XXX: This code is not updated with git repository

# Prepare data for analysis
df['property_tax_rate'] = df['property_tax'].apply(lambda r: r['rate'])
df['property_tax_value'] = df['property_tax'].apply(lambda r: r['value'])
df['property_tax_value_per_person'] = df['property_tax_value']/df['population']
df['home_tax_rate'] = df['home_tax'].apply(lambda r: r['rate'])
df['home_tax_value'] = df['home_tax'].apply(lambda r: r['value'])
df['home_tax_value_per_person'] = df['home_tax_value']/df['population']
df['debt_ratio'] = df['debt_annual_costs']/df['operating_revenues']
df['staff_costs_ratio'] = df['staff_costs']/df['operating_revenues']
df['staff_costs_per_person'] = df['staff_costs']/df['population']
```

### Répartition de la taxe d'habitation et de la taxe foncière

On notera que la taxe foncière et la taxe d'habitation sont en moyenne très proche, de l'ordre de 13%, avec un écart-type un peu plus important sur la taxe foncière.

```python
df[['property_tax_rate', 'home_tax_rate']].describe()
```

Une disparité entre les communes qui peut être très importante: on passe de 0% à 59% pour la taxe foncière, de 0% à 61% pour la taxe d'habitation.

```python
plt.figure(figsize=(12,12));
df[['property_tax_rate', 'home_tax_rate']].boxplot()
df[['property_tax_rate', 'home_tax_rate', 'name', 'insee_code']].head(n=20)
```

```python
# Biggest property tax rate
_df = df.sort(columns='property_tax_rate', ascending=False)
_df[['property_tax_rate', 'property_tax_value', 'name', 'insee_code']].head(n=20)
```
